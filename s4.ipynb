```python
data_sort_z=data.sort_values('photoz',ascending=False)

ccc=int(len(data_sort_z)/5)

data_z_3=data_sort_z.iloc[0:ccc,:]
data_z_2=data_sort_z.iloc[2*ccc:3*ccc,:]
data_z_1=data_sort_z.iloc[-ccc:,:]
```
